## Step 4: Apply the Analytics Lifecycle

**Problem Statement:** Does money buy happiness? (GDP vs Happiness analysis)


```python
# DESCRIPTIVE ANALYTICS: What are the happiness scores by region?
regional_stats = df.groupby('region').agg(
    avg_happiness = ('happiness_score', 'mean'),
    avg_gdp = ('gdp_per_capita', 'mean'),
    country_count = ('country', 'count')
).round(2).sort_values('avg_happiness', ascending=False)

print('=== HAPPINESS BY REGION (Descriptive Analytics) ===')
display(regional_stats)
```

```python
# DIAGNOSTIC ANALYTICS: Which factors correlate most with happiness?
numeric_cols = ['gdp_per_capita', 'social_support', 'life_expectancy',
                'freedom_score', 'corruption_perception']

correlations = df[numeric_cols + ['happiness_score']].corr()['happiness_score'].drop('happiness_score')
correlations_sorted = correlations.abs().sort_values(ascending=False)

print('=== FACTORS CORRELATED WITH HAPPINESS (Diagnostic Analytics) ===')
for factor, corr in correlations.items():
    direction = '📈 Positive' if corr > 0 else '📉 Negative'
    print(f'  {factor:30s}: {corr:+.3f}  {direction}')
```

```python
# VISUALIZATION: GDP vs Happiness scatter plot
plt.figure(figsize=(12, 7))

# Color points by region
regions = df['region'].unique()
colors = plt.cm.Set2(range(len(regions)))
color_map = dict(zip(regions, colors))

for region in regions:
    mask = df['region'] == region
    plt.scatter(df[mask]['gdp_per_capita']/1000,
                df[mask]['happiness_score'],
                label=region, s=80, alpha=0.8,
                color=color_map[region])

# Add country labels for notable points
for _, row in df.iterrows():
    if row['happiness_score'] > 7.3 or row['happiness_score'] < 4.0 or row['country'] == 'India':
        plt.annotate(row['country'],
                     (row['gdp_per_capita']/1000, row['happiness_score']),
                     textcoords='offset points', xytext=(5, 5), fontsize=8)

plt.xlabel('GDP per Capita (in $000s)', fontsize=12)
plt.ylabel('Happiness Score (0-10)', fontsize=12)
plt.title('Does Money Buy Happiness?\nGDP per Capita vs. Happiness Score by Country', fontsize=14)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
print('\n💡 Insight: There IS a positive correlation between GDP and happiness, but diminishing returns above ~$50K!')
```
